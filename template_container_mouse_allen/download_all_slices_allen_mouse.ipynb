```python
#download mouse slices
import os
import requests
from PIL import Image, ImageDraw, ImageColor, ImageFilter, ImageFont
dirpath = os.getcwd()
if os.path.exists(dirpath+'/slice_numbers')==False:
    os.makedirs(dirpath+'/slice_numbers')
blank = 'https://scalablebrainatlas.incf.org/templates/ABA_v3/coronal_STP_1675/P56_Atlas_0000.jpg'
# the slices # correspond to those in parentheses on https://scalablebrainatlas.incf.org/mouse/ABA_v3
#the first slice is 2
#the last slice 526
# each 4 slice to be downloaded
i = 526
for k in range(0, 132):
    temp = blank[0:(len(blank) - (4+len(str(i))))]
    url = temp+str(i)+'.jpg'
    r = requests.get(url)
    with open(str(dirpath+'/slice_numbers'+'/'+str(i)+'.png'),'wb') as f: 
        f.write(r.content)
    i =i-4
```

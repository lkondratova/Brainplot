```python
#download rat slices. Rat - MRH Atlas of the developing brain: P80
import os
import requests
from PIL import Image, ImageDraw, ImageColor, ImageFilter, ImageFont
dirpath = os.getcwd()
if not os.path.exists(dirpath+'/slice_numbers'):
    os.makedirs(dirpath+'/slice_numbers')
blank = 'https://scalablebrainatlas.incf.org/templates/CBWJ13_age_P80/coronal_GRE/p80_average_gre_'
# the slices # correspond to those in parentheses on https://scalablebrainatlas.incf.org/rat/CBWJ13_age_P80
#the first slice is 84
#the last slice 1424
#134 slices in total
# each 10th slice to be downloaded
i = 1424
for k in range(0, 134):
    url = blank+str(i)+'.jpg'
    r = requests.get(url)
    with open(str(dirpath+'/slice_numbers'+'/'+str(i)+'.png'),'wb') as f: 
        f.write(r.content)
    #img=Image.open(dirpath+'/slice_numbers'+'/'+str(i)+'.png')
    #im=img.convert('RGBA')
    #im.save(dirpath+'/slice_numbers'+'/'+str(i)+'.png','PNG')
    i =i-10
```
